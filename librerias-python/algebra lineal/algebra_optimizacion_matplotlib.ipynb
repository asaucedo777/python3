```python
"""
Planteamiento del problema:
  Dada f(x) = 50*x1 + 40*x2
    x1 = -40*x2 / 50
    x2 = -50*x1 / 40
  Y las restricciones:
    x1 + 1.5*x2 <= 750
    2*x1 + x2 <= 1000
    x1 >= 0
    x2 >= 0
  Obtener el valor máximo de x en el intervalo (0,oo)
"""
```

```python
# Importart numpy y matplotlib.pyplot
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definimos el gráfico
x = np.linspace(0, 1000, 10)   # Los valores de x1 y x2 están entre 0 y 1000 (máximo valor que se hace 0)
plt.plot(x, (750 - x) / 1.5)  # Según restricción x2 <= (750-x1) / 1.5  -> se hace 0 en x = 750
plt.plot(x, 1000 - 2*x)       # Según restricción x2 <= (1000-2*x1)     -> se hace 0 en x = 1000
plt.axis(ymin=0)
plt.grid(True)
plt.show()
```
