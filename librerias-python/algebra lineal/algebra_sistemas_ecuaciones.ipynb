```python
# Importart numpy y matplotlib.pyplot
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Plotear varias variables con en un único plt.plot (KO)
eje_x = np.arange(0,10,0.1)
plt.plot(
  eje_x,   # Eje x (50 valores separados por 0.1)
  (1 - eje_x)/-2,      # y0
  (11 - (3*eje_x))/2,  # y1
)
```

```python
# Plotear variables con varios plt.plot (OK)
x_vals = np.linspace(start=0.1, stop=5, num=50)
plt.plot(x_vals, (1 - x_vals)/-2)
plt.plot(x_vals, (11 - (3*x_vals))/2)
plt.axis(ymin=0)
```

```python
# Ejemplo plotear sistema ecuaciones
"""Plotear el sistema de ecuaciones:
  x  -2y = 1      y = f(x) = (1-x)/-2
  3x +2y = 11     y = f(x) = (11-3x)/11
"""
x = np.arange(0, 5, 0.1)
plt.plot(x, (1-x)/-2)
plt.plot(x, (11-3*x)/11)
```

```python
# Representar sistema de ecuaciones con matrices
""" 
  x  -2y = 1      1 -2   x   1
  3x +2y = 11     3  2   y  11
"""
A = np.array([
  [1., -2.],
  [3., 2.]
])
C = np.array([1., 11.])
print(A)
print(C)
```

```python
# Resolver sistema de ecuaciones utilizando la función solve
X = np.linalg.solve(A, C)
print(X)
# Producto escalar de las matrices
print(A @ X)
```
