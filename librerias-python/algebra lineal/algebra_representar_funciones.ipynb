2. Representación de f(x) = x2
3. Representación de f(x) = 1/x

```python
# Importart numpy y matplotlib.pyplot
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Plotear una función
x = np.arange(-10, 10, 0.1)
plt.plot(x**2)
```

```python
# Representación de f(x) = x2
x = np.arange(-10, 10, 0.1)
fig, ax = plt.subplots()
ax.grid()
plt.plot(x, x**2)
```

```python
# Representación de f(x) = (x2 - 2x) / (x2 + 2x)
x_i = np.arange(-1, 0.1, 0.01)
x_d = np.arange( 0.1, 1, 0.01)
fig, ax = plt.subplots()
ax.grid()
plt.plot(x_i, (x_i**2 - 2*x_i) / (x_i**2 + 2*x_i))
plt.plot(x_d, (x_d**2 - 2*x_d) / (x_d**2 + 2*x_d))
```

```python
# Representación de f(x) = 1/x
x_izda =  np.arange(-1, 0, 0.01)  # para x < 0
x_drcha = np.arange( 0, 1, 0.01) # para x > 0
fig, ax = plt.subplots()
ax.grid()
plt.plot(x_izda, 1/x_izda)
plt.plot(x_drcha, -1/x_drcha)
```
