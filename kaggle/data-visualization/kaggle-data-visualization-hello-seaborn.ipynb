```python
# Exercise
museum_data = pd.read_csv('../../../data/museum_visitors.csv', index_col='Date', parse_dates=True)
museum_data.head()
```

```python
# How many visitors did the Chinese American Museum receive in July 2018?
museum_data.loc['2018-07-01']['Chinese American Museum']
```

```python
# In October 2018, how many more visitors did Avila Adobe receive than the Firehouse Museum?
museum_data['Avila Adobe'].loc['2018-10-01':'2018-10-31'].sum() - museum_data['Firehouse Museum'].loc['2018-10-01':'2018-10-31'].sum() 
```

```python
# plot
plt.figure(figsize=(16,6))
sns.lineplot(museum_data)
```

```python
# plot only Avila Adobe
sns.lineplot(museum_data['Avila Adobe'])
```

```python
sns.lineplot(museum_data['Avila Adobe'].loc['2014-01-01':'2015-01-01'])
```

```python
"""
Does Avila Adobe get more visitors:

in September-February (in LA, the fall and winter months), or
in March-August (in LA, the spring and summer)?
"""
sep_feb = museum_data['Avila Adobe'].loc['2014-09-01':'2015-02-01'].sum()
mar_aug = museum_data['Avila Adobe'].loc['2015-03-01':'2015-08-01'].sum()
sep_feb, mar_aug
```

```python
# plot bar char
plt.figure(figsize=(16,6))
sns.barplot(museum_data['Avila Adobe'])
```

```python
# Heatmaps
sns.heatmap(museum_data)
```
