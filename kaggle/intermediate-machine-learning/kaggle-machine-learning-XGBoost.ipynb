**XGBoost**

```python
# load data
import pandas as pd
data = pd.read_csv("../../../data/melb_data.csv")
# Select subset of predictors
cols_to_use = ['Rooms', 'Distance', 'Landsize', 'BuildingArea', 'YearBuilt']
X = data[cols_to_use]
y = data.Price
```

```python
# separate train and validation sets
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(X, y)
```

```python
# import XGBoost regressor
from xgboost import XGBRegressor
model = XGBRegressor()
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error
predictions = model.predict(X_valid)
print(f"MAE {mean_absolute_error(predictions, y_valid):.2f}")
```

```python
# adjust parameters: n_estimators
model_estim = XGBRegressor(n_estimators=500)
model_estim.fit(X_train, y_train)
predictions = model_estim.predict(X_valid)
print(f"MAE {mean_absolute_error(predictions, y_valid):.2f}")
```

```python
# adjust fit parameters: early_stopping_rounds
model_early = XGBRegressor(n_estimators=500)
model_early.fit(X_train, y_train, 
             early_stopping_rounds=5, 
             eval_set=[(X_valid, y_valid)],
             verbose=False)
predictions = model_early.predict(X_valid)
print(f"MAE {mean_absolute_error(predictions, y_valid):.2f}")
```

```python
# adjust learning_rate
model_learn = XGBRegressor(n_estimators=1000, learning_rate=0.05)
model_learn.fit(X_train, y_train, 
             early_stopping_rounds=5, 
             eval_set=[(X_valid, y_valid)], 
             verbose=False)
predictions = model_learn.predict(X_valid)
print(f"MAE {mean_absolute_error(predictions, y_valid):.2f}")
```
